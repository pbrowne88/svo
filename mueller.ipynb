# NLP and SVO

The first thing we will do is get the raw text out of the `html` markup and get rid of newline characters. Once we do that, we will feed each paragraph through `spaCy`. 

```python
for_proc = core_text
for_proc = [str(p.text.replace('\n',' ')) for p in for_proc] 
```

One common problem in this data -- based on just some preliminary analyses -- is that footnote ids are attached to punctuation, which `spaCy` seems to trip up on as it thinks it is one weird string, like 'campaign.476' rather than three different tokens. This is intended to quickly fix that up. See screenshot "the_number_problem.png". Note that this is probably *not* the best solution. It might be better to leave the pattern there but pad it with white space. 

```python
pattern_find = r'[0-9]+' # any numbers 
pattern_replace = '' # replace them with nothing 
for_proc = [re.sub(pattern_find, pattern_replace, t) for t in for_proc]
```

# To Narrative Networks

We can easily get this data into a network structure, following Franzosi, Bearman and Stovel, and others. 

The semantic triples above contain `subject` ($s$), `verb` ($v$), `object` ($o$). We want to construct a network where nodes are subjects and objects (we will assume they are of the same node mode), and the links are (a) directed from $s$ to $o$, and (b) labeled by the verb. 

So I am going to create a `dict` where the $s$,$o$ tuple is a key and the `v` is the value. This can then be used to provide a number of useful things to `networkx`, including a dictionary of edge labels that can be used to plot parts of the semantic network. 

```python
svos = {}
for para in report_svos:
    for t in para:
        if len(t) > 1:
            sub = str(t[0]).replace(" ","_")
            obj = str(t[2]).replace(" ","_")
            etuple = (sub, obj)
            svos[etuple] = str(t[1]).lower().replace(" ","_") # this is where we assign the verb as the value for the tuple key
```

```python
svos
```

We now have a dictionary where (1) the `keys` are a tuple of `s` and `o`, and the `values` are the label for the verb that connected $s$ to $o$.

```python
counts = dict(Counter([item for sublist in report_svos for item in sublist]))
weights = {}
for k,v in counts.items():
    sub = str(k[0]).replace(" ","_")
    obj = str(k[2]).replace(" ","_")
    weights[(sub,obj)] = v
```

From here we can pull a list of edges out from the keys and pass it to `networkx`. 

```python
G = nx.DiGraph()
G.add_edges_from([k for k in svos.keys()])
G.remove_edges_from(nx.selfloop_edges(G)) # add edge labels to the network. since this is the full network it will accept SVO master list. 
nx.set_edge_attributes(G, svos, 'actions')
nx.set_edge_attributes(G, weights, 'weight')
G.remove_edges_from(nx.selfloop_edges(G))
G.name = 'Volume 1 SemNet'
print(nx.info(G))
```

For the full report of Volume 1, we have over 4K nodes and over 4K edges. We should expect these numbers to be similar. SVOs operationalized this way are sparse, so each one is likely to become it's own unique $s$ and $o$ nodes linked by a unique $v$. But if the same SVO does show up more than once, then the number of edges will drop slightly by the number of nodes will not. We can pull out ego networks for specific nodes. Again, thinking in terms of the pronoun issue, we should not expect 'Manafort' to return everything related to Manafort, because some might have 'Manafort' as part of a larger string, or perhaps uses a pronoun, etc.

```python
def get_egos(network, so_search_term, print_overview=True):
    """
    The search term is case sensitive. 
    
    Two objects are returned. The first is a single network consisting of the
    composed galaxy of ego networks for all nodes that include the search 
    string in their node label. The second object is a list of the uncomposed networks.
    
    TODO 
    ----
    
    If mode is 'merge', then it merges all the nodes into one node and returns the ego 
    network from the new node in the entire original SVO network. 
    
    If mode is 'galaxy' then it composes them into one network. Not one node.
    
    For merge? https://networkx.github.io/documentation/stable/reference/algorithms/generated/networkx.algorithms.minors.contracted_nodes.html
    """
    so_search_term_nodes = [n for n in network.nodes() if so_search_term in n]
    
    # get the ego nets
    egos = []
    egos_vs = []
    for node in so_search_term_nodes:
        ego = nx.ego_graph(network, node)
        ego.name = node
        egos.append(ego)
        
    # compose the gallaxy network 
    net = nx.DiGraph()
    for e in egos:
        net = nx.compose(net, e)
    
    
    # print the gallaxy descriptives 
    if print_overview is True:
        info = []

        for ego in egos:
            i = {}
            i['nnodes'] = nx.number_of_nodes(ego)
            i['nedges'] = nx.number_of_edges(ego)
            i['name'] = ego.name
            info.append(i)
        
        info = sorted(info, key=itemgetter('nedges'), reverse=True)
        print('NN ', 'NE ', 'Network Name / Node Label')
        print(tabulate(info))

    return net, egos, egos_vs
```

```python
one, multiple, ego_verbs = get_egos(G, 'Trump')
```

Given a list of networks like this, we might construct several criteria for the ego networks we want to merge into our gallaxy network. So, for example, we might say compose all networks that have more than 4 edges, or compose all networks in this list... And so on. So better not to write a method for this or anything like that. 

```python
a = nx.DiGraph()

for net in multiple:
    if net.number_of_edges() > 3:
        a = nx.compose(a, net)

a.name = 'Curated Trump'
print(nx.info(a))
```

```python
actions = {}
for e in a.edges(data=True):
    actions[(e[0], e[1])] = e[2]['actions']


pos = nx.kamada_kawai_layout(a)
nx.draw(a, pos=pos,with_labels = True, node_color = 'lightgray', edge_color = 'darkgray', font_size = 8, node_size = 200)
el = nx.draw_networkx_edge_labels(a, pos, edge_labels=actions, font_size = 4, font_color='red') 
```
